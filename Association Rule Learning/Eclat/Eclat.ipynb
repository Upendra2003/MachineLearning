```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
#Appending every row and column element in a list transactions
transactions=[]
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])
# we have make a list of transactions and make every element as a string
```

```python
from apyori import apriori
rules = apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=1,max_length=3)
# min_support = 3*7/7501 this means that a min of 3 items in a day are sold
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results] 
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```
